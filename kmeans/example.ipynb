```python
# K-Means Clustering Example: Grouping Data Points by Similarity üîç

# Import necessary libraries
# -----------------------------------------------
# numpy: For numerical operations and array handling
# matplotlib.pyplot: For visualizing clusters
# sklearn.datasets.make_blobs: To generate synthetic clustered data
# kmeans: Our custom module containing the KMeansModel class

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from kmeans import KMeansModel

```

```python
# ---------------------------------------------------------
# Step 1: Generate synthetic data with 3 clusters
# ---------------------------------------------------------

# We use sklearn's make_blobs to create sample data points grouped into 3 clusters
# n_samples=300: total number of points
# centers=3: number of clusters
# cluster_std=1.0: spread of clusters
# random_state=42: reproducibility of results

X, y_true = make_blobs(n_samples=300, centers=8, cluster_std=1.0, random_state=42)
```

```python

# ---------------------------------------------------------
# Step 2: Create and train the KMeans model
# ---------------------------------------------------------

# Initialize our KMeansModel with 3 clusters (k=3)
model = KMeansModel(n_clusters=5, max_iters=100)

# Fit the model on our data
# This assigns each point to a cluster and updates cluster centers iteratively
model.fit(X)

```

```python
# ---------------------------------------------------------
# Step 3: Get the predicted cluster labels and centers
# ---------------------------------------------------------

labels = model.predict(X)  # cluster assignments for each point
centers = model.centers   # coordinates of cluster centers

```

```python
# ---------------------------------------------------------
# Step 4: Visualize the clusters and their centers
# ---------------------------------------------------------

plt.figure(figsize=(10, 7))

# Plot each cluster with different colors based on labels
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', marker='o', alpha=0.6, label='Data points')

# Plot the cluster centers with big red stars
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='*', s=300, label='Cluster Centers')

# Add titles and labels for clarity
plt.title('K-Means Clustering Result')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()
```
